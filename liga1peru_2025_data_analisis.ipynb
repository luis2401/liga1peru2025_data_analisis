```python
import numpy as np
import pandas as pd
import matplotlib as plt
import unidecode as uni
```

```python
#Extración de datos
df = pd.read_csv("data/liga1_peru.csv")


df.info()
#Limpieza de datos
df["Home_Team"] = (
    df["Home_Team"]
    .str.normalize("NFKD")
    .str.encode("ascii", errors="ignore")
    .str.decode("utf-8")
)
df["Home_Team"] = df["Home_Team"].apply(uni.unidecode)
df["Away_Team"]=df["Away_Team"].fillna("Unknowm")

df["Home_Goals"]=df["Home_Goals"].replace("two", "2")
df["Home_Goals"]=df["Home_Goals"].replace("one", "1")
df["Home_Goals"]=df["Home_Goals"].astype(int)




df.head(40)

```
